```python
import kagglehub

path = kagglehub.dataset_download("annavictoria/speed-dating-experiment")
```

```python
import pandas as pd
file_path = path + "/Speed Dating Data.csv"
print(file_path)
df = pd.read_csv(file_path, encoding='latin-1')
df.head()
```

```python
#Time1
columns_Time1 = ['iid', 'pid']
for col in df.columns:
    if col.endswith('1_1'):
        columns_Time1.append(col)
df_pre_event = df[columns_Time1]

#Scorecard
columns_scorecard = ['iid', 'pid', 'attr', 'sinc', 'intel', 'fun', 'amb', 'shar', 'dec', 'match']
df_scorecard = df[columns_scorecard]

#Half-way
columns_Halfway = ['iid', 'pid']
for col in df.columns:
    if col.endswith('1_s'):
        columns_Halfway.append(col)
df_halfway = df[columns_Halfway]

#Time2
columns_Time2 = ['iid', 'pid']
for col in df.columns:
    if col.endswith('1_2'):
        columns_Time2.append(col)
df_day_after = df[columns_Time2]

#Time3
columns_Time3 = ['iid', 'pid']
for col in df.columns:
    if col.endswith('1_3'):
        columns_Time3.append(col)
df_weeks_later = df[columns_Time3]


print("Pre-Event Survey Data ---")
display(df_pre_event.head(3))

print("\nIn-Event Scorecard (The Date) ---")
display(df_scorecard.head(3))

print("\nHalfway Survey Data ---")
display(df_halfway.head(3))

print("\nDay-After Survey Data ---")
display(df_day_after.head(3))

print("\n3-4 Weeks Later Survey Data ---")
display(df_weeks_later.head(3))
```
