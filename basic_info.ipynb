```python
import kagglehub

path = kagglehub.dataset_download("annavictoria/speed-dating-experiment")
```

```python
import pandas as pd
file_path = path + "/Speed Dating Data.csv"
print(file_path)
df = pd.read_csv(file_path, encoding='latin-1')
df.head()
```

```python
print("Dataset Shape:", df.shape)

df.info(verbose=True, show_counts=True)
```

```python
print(df.describe().T)
```

```python
missing_percentages = (df.isnull().sum() / len(df)) * 100
missing_data = missing_percentages[missing_percentages > 0].sort_values(ascending=False)

print("Columns with most missing data:")
print(missing_data.head(10))
```

```python
print("Total Match Count:")
print(df['match'].value_counts())

print("\nMatch Percentage:")
print(df['match'].value_counts(normalize=True) * 100)
```
